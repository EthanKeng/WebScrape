```python
import requests
from bs4 import BeautifulSoup
import csv
import random,time
# import json
# import pandas as pd
# import os

# if os.path.exists('./wear') == False:
#     os.mkdir('./wear')


url = 'https://www.biccamera.com/bc/category/001/100/009/005/'
headers = {
"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36",
'Referer': '{}'.format(url)
}

referer =  ["https://www.google.com/","https://www.google.co.uk","biccamera.rakuten.co.jp","playstation.com","https://www.biccamera.com/","https://www.google.co.uk","biccamera.rakuten.co.jp","playstation.com","https://www.biccamera.com/","https://www.google.co.uk","biccamera.rakuten.co.jp","playstation.com","https://www.biccamera.com/","https://www.google.co.uk","biccamera.rakuten.co.jp","playstation.com","https://www.biccamera.com/"]
pageRef = 0
productUrl_list = []

for page in range(160):
    nextPage = url+"?p={}".format(page+1)
    res = requests.get(nextPage, headers=headers)
    soup = BeautifulSoup(res.content, 'html.parser')
    itemLinks = soup.find_all('p', class_='bcs_title')
    for title in soup.find_all('title'):
        print(title.get_text())
        time.sleep(5)
    if (page+1) % 10 == 0:
        time.sleep(60)
        headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36",
        'Referer': '{}'.format(referer[pageRef])
        }
        pageRef += 1

    for i in itemLinks:
        itemUrl = i.find('a').get('href')
        itemTitle = i.text.strip()

        if "biccamera.com" not in itemUrl:
            continue
        productUrl_list.append([itemTitle,itemUrl])
        
```

```python
import csv
file = open("urls.csv","w+",newline='')
with file:
    write = csv.writer(file)
    write.writerows(productUrl_list)
    
```
